### Series

```python
# los índices pueden ser de más tipos
football_cities = pd.Series(['Barcelona','Madrid','Valencia','Sevilla'],
                            index=['a','b','c','d'])
print(football_cities)
```

```python
# uso de masks para seleccionar
fibonacci = pd.Series([0, 1, 1, 2, 3, 5, 8, 13, 21])
print(fibonacci)

mask = fibonacci > 10
print(mask)
print(fibonacci[mask])

dst = pd.Series([13,21])
print(dst)

dst.equals(fibonacci)

fb = fibonacci[mask]
fb.reset_index(drop=True, inplace=True)
print(fb)

dst.equals(fb)
```

#### 3.2 Operacion entre series

```python
# suma de dos series
# suma de valores con el mismo índice (NaN si no aparece en ambas)
serie1 = pd.Series([10,20,30,40], index=range(4) )
serie2 = pd.Series([1,2,3], index=range(3) )
suma = serie1 + serie2
print(suma)
```

```python
# resta de series (similar a la suma)
print(serie1- serie2)
```

```python
# operaciones de pre-filtrado
result = serie1 + serie2
result[pd.isnull(result)] = 0 # mask con isnull()
print(result)
```

#### 3.2.1 Diferencias entre Pandas Series y diccionario
* Diccionario, es una estructura que relaciona las claves y los valores de forma arbitraria.
* Series, estructura de forma estricta listas de valores con listas de índice asignado en la posición.
* Series, es más eficiente para ciertas operaciones que los dicionarios.
* Enlas Series los valores de entrada pueden ser listas o Numpy arrays.
* EnSeries los índices semánticos pueden ser integers o caracteres, en los valores igual.
* Series se podría entender entre una lista y un diccionario Python, pero es de una dimensión.

### 4.3 Indexación y slicing con DataFrames

```python
d1 = {'ciudad':'Valencia', 'temperatura':10, 'o2':1}
d2 = {'ciudad':'Barcelona', 'temperatura':8}
d3 = {'ciudad':'Valencia', 'temperatura':9}
d4 = {'ciudad':'Madrid', 'temperatura':10, 'humedad':80}
d5 = {'ciudad':'Sevilla', 'temperatura':15, 'humedad':50, 'co2':6}
d6 = {'ciudad':'Valencia', 'temperatura':10, 'humedad':90, 'co2':10}

ls_data = [d1, d2, d3, d4, d5, d6] # lista de diccionarios
df_data = pd.DataFrame(ls_data, index = list('abcdef'))
display(df_data)
```

```python
# Acceso a un valor concreto por indice posicional [row, col]
print(df_data.iloc[1,1])
# Acceso a todos los valores hasta un índice por enteros
display(df_data.iloc[:3,:4])
# Acceso a datos de manera explícita, indice semantico (se incluyen)
display(df_data.loc['a', 'temperatura'])
display(df_data.loc[:'c', :'o2'])
display(df_data.loc[:'c', 'temperatura':'o2'])
display(df_data.loc[:, ['ciudad','o2']])
```

```python
# indexación con nombre de columna (por columnas)
print(df_data['ciudad']) #--> Series
display(df_data[['ciudad', 'o2']])
```

```python
# indexación con índice posicional (no permitido!). Esto busca columna.
df_data.iloc[0]
```

```python
# indexar por posición con 'iloc'
print(df_data.iloc[0]) #--> Series de la primera fila (qué marca los índices)
```

```python
# indexar semántico con 'loc'
df_data.loc['a'] #--> Series de la fila con índice 'a'
```

```python
# indexar semántico con 'loc'
df_data.loc[:'b'] #--> DataFrame de la fila con índice 'a' y 'b'
```

```python
df_data.loc[:'b'].loc[:,["o2", "humedad"]] # slicing anidado
```

```python
# si se modifica una porcion del dataframe se modifica el dataframe original␣(referencia)
display(df_data)

serie = df_data.loc['a']
print(serie)
serie.iloc[2] = 3000 # setting with copy warning!!!
display(df_data)

# copiar data frame
df_2 = df_data.loc['a'].copy()
df_2.iloc[2] = 3000
display(df_2)
display(df_data)
```

```python
# ambos aceptan 'axis' como argumento
# df_data.iloc(axis=1)[0] #--> todos los valores asignados a la primera columna 'ciudad'
df_data.loc(axis=1)['ciudad'] #--> equivalente frame['ciudad']
```

* Hacer un análisis general de los otros dos archivos CSV en ml-1m (‘movies.dat’ y ‘ratings.dat’)

```python
import pandas as pd
```

```python
#movies.dat

import pandas as pd

path = "res/ml-1m/"

# Carga de datos
movies = pd.read_csv(f"{path}movies.dat", sep='::', header=None, 
                     names=['MovieID', 'Title', 'Genres'], engine='python', encoding='latin-1')

print("--- ANÁLISIS DE: MOVIES.DAT ---")
# 1. Dimensiones
print(f"Dimensiones: {movies.shape[0]} películas con {movies.shape[1]} características.")

# 2. Integridad de datos
print(f"Valores nulos por columna:\n{movies.isnull().sum()}")

# 3. Análisis de Contenido (Géneros)
# Separamos los géneros ya que vienen agrupados por '|'
generos_count = movies['Genres'].str.split('|', expand=True).stack().value_counts()
print(f"\nTotal de géneros únicos encontrados: {len(generos_count)}")
print("Top 5 géneros con más títulos:")
print(generos_count.head(5))

# 4. Muestra de datos
print("\nMuestra del catálogo:")
print(movies[['Title', 'Genres']].head())
```

```python
# ratings.dat

# Carga de datos
ratings = pd.read_csv(f"{path}ratings.dat", sep='::', header=None, 
                      names=['UserID', 'MovieID', 'Rating', 'Timestamp'], engine='python')

print("\n--- ANÁLISIS GENERAL: RATINGS.DAT ---")
# 1. Volumen de transacciones
print(f"Total de calificaciones registradas: {len(ratings)}")

# 2. Análisis Estadístico del Rating
print("\nResumen descriptivo de las calificaciones:")
print(ratings['Rating'].describe())

# 3. Distribución de Calificaciones
print("\nConteo de votos por estrella (1-5):")
distribucion = ratings['Rating'].value_counts(normalize=True).sort_index() * 100
for estrella, porcentaje in distribucion.items():
    print(f"Nota {estrella}: {porcentaje:.2f}% de los votos")

# 4. Actividad de Usuarios
print(f"\nUsuarios únicos: {ratings['UserID'].nunique()}")
print(f"Promedio de películas calificadas por usuario: {len(ratings)/ratings['UserID'].nunique():.2f}")
```

* Analizando el dataset ratings.dat, ¿hay algún usuario que no tenga ninguna review? ¿Cuántos tienen menos de 30 reviews?

```python
# Contar cuántas reviews tiene cada usuario
reviews_por_usuario = ratings.groupby('UserID').size()

# Verificar el valor mínimo de reviews realizadas por un usuario
min_reviews = reviews_por_usuario.min()

print(f"El número mínimo de reviews por usuario en el archivo es: {min_reviews}")

if min_reviews > 0:
    print("No hay usuarios en este archivo con 0 reviews; todos tienen al menos una.")
else:
    print("Existen usuarios sin reviews registradas.")
```

```python
# Filtrar los usuarios que tienen menos de 30 registros
usuarios_menos_30 = reviews_por_usuario[reviews_por_usuario < 30]

# Contar cuántos cumplen la condición
cantidad_usuarios = len(usuarios_menos_30)

print(f"Cantidad de usuarios con menos de 30 reviews: {cantidad_usuarios}")

# Opcional: Mostrar los primeros 5 usuarios que cumplen esto y su cantidad exacta
print("\nEjemplo de usuarios con poca actividad:")
print(usuarios_menos_30.head())
```
