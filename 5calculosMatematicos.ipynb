```python
aa = int(input('Ingresar año actual: '))
an = int(input('Ingresar año de nacimineto: '))

ma = int(input('Ingresar mes actual: '))
mn = int(input('Ingresar mes de nacimiento: '))

da = int(input('Ingresar dia actual: '))
dn = int(input('Ingresar dia de nacimiento: '))

años = aa-an
print ('Edad en años: ', años)

meses = (aa-an) * 12 + (ma-mn)
print ('Edad en meses: ', meses)

dias = (aa-an) * 365 + (ma-mn) * 30 + (da-dn)
print ('Edad en dias: ', dias)
```

3.- Elaborar un programa en python que permite ingresar el radio y la altura de un cilindro y calcular el volumen del cilindro sabiendo que la formula es pi.r2.h

```python
radio = float(input('Ingresar el radio: ',))
altura = float(input('Ingresar la altura: ',))
pi = 3.1416
print ('El valor de pi es de: ', pi)
volumen = pi * (radio ** 2) * altura
print ('El calculo matematico es: ', volumen)
```

4.- Elaborar un programa que permita convertir una distancia medida en metros a centimetros, milímetros y kilómetros. El usuario debe ingresar el valor en metros y el programa debe mostrar las tres equivalencias correspondientes.

```python
distancia = float(input('Ingresar la distancia en metros: '))

metro = ( distancia * 100 )
print ('Conversion en centimetros: ', metro)

milimetro = (distancia * 1000)
print ('Conversion en milimetros: ', milimetro)

Kilometros = ( distancia / 1000 )
print ('Conversion en kilometros: ', Kilometros)
```
