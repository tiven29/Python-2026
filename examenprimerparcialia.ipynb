$$IMC=\frac{Peso}{estatura^2}$$
	​


```python
peso = int(input("Ingresar peso kg: ")) ## Asignamos una varible para guardar el peso
estatura = float(input("Estatura: ")) ##cremamos la varible estatura para almacenar los datos a ingresar con float 
IMC = (peso/estatura**2) ##realizamos el calculculo para ver 
print ('Su resultado es: ', IMC) ##presentamos el resultado de los datos ingresados 
```
