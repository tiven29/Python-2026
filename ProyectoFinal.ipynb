```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Añadimos sep=';' para que reconozca las columnas correctamente
df = pd.read_csv('res/vgsales.csv')

# Ahora sí, el dataset se verá ordenado como en Excel
df.head()
```

```python
# El parámetro include='all' asegura que veas tanto columnas de texto como numéricas
display(df.describe(include='all'))
```

```python
# 1. Conteo de nulos por columna
print("--- Cantidad de valores nulos por columna ---")
nulos = df.isnull().sum()
print(nulos)

# 2. Cálculo del porcentaje de "completitud" (como en tu imagen de referencia)
completo = (df.notna().mean() * 100)
print("\n--- Porcentaje de datos presentes ---")
for col, valor in completo.items():
    print(f"Completitud de {col}: {valor:.2f}%")
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Mapa de calor de datos nulos')
plt.show()
```
