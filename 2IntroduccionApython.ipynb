###### **3.0.12 Performance Sample, NumPy vs Plain Python**

```python
import numpy as np
my_arr = np.arange(10000000) # Utilizando NumPy Arrays
my_list = list(range(10000000)) # Utilizando Listas
print(my_list)
```
