2.- Ingresar por teclado la fecha actual dia, mes, año actual, y tambien ingresar la fecha de nacimiento de una persona y calcular cuantos años tiene esa persona en dias mes y año

```python
aa = int(input("Ingresar año actual: "))
ma = int(input("Ingresar mes actual: "))
da = int(input("Ingresar dia actual: "))

an = int(input("Ingresar año de nacimiento: "))
mn = int(input("Ingresar mes de nacimiento: "))
dn = int(input("Ingresar dia de nacimiento: "))

años = aa - an
print('edad en años: ', años )
meses_totales = (aa - an) * 12 + (ma - mn)
print('edad en meses totales: ', meses_totales)
dias_totales = (aa - an) * 365 + (ma - mn) * 30 + (da - dn)
print('edad en dias totales: ', dias_totales)
```
