1. Realizar un programa en Python donde permita al usuario ingresar un numero entero positivo repititivamente el programa debe continuar mientras se ingrese un numero mayor a cero. Al finalizar el proceso debe mostrarne la suma de los nimeros ingresados

```python
suma = 0
a = int(input("Ingresa el primer numero: "))
while a > 0:
    suma = suma + a
    a = int(input("Ingresa otro número: "))
print(f"La suma total de los números (positivos) ingresados es: {suma}")
```

2. Crear un programa en Pyhon donde el usuario ingrese un numero por teclado verificar si un numero es primo o no utilizando while

```python
n = int(input("Ingrese un número: "))
if n < 2:
    print("no primo")
else:
    i = 2
    while i*i <= n:
        if n % i == 0:
            print("es número no primo")
            break
        i += 1
    else:
        print("es número primo")
```

3. Desarrollar un programa en Python donde el usuario ingrese dos numeros por teclado, el primer numero va a ser inicio de un rango ,el segundo va a ser el final de dicho rango mediante el ciclo while verificar cuantos numeros son pares y mostrar los numeros y total de los numeros pares.

```python
suma = 0
contador = 0
while True:
    numero = float(input("Ingrese un número: "))
    if numero < 0:
        break
    suma = suma + numero
    contador = contador + 1
if contador > 0:
        print(f"Se ingresaron {contador} números positivos")
        print(f"La suma total de números los números positivos es: {suma}")
else:
        print("No se ingresó ningún número positivo")
```
