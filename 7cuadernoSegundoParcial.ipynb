### 1 Estructuras de control
#### 1.1 Seleccion (sentencias condicionales)
* Seleccion de una de varias alternativas en base a alguna condicion
* Indentación para estructurar el código
* Importante el símbolo ‘:’

```python
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente

```python
x = 3
y = [1, 2, 3]

print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

```python
#Desarrolar un programa en python donde se ingrese un numero verificar si el 
#numero ingresado es positivo, negativo o nulo
x =-3
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

### Valo Booleano

* Todos los objetos en python tienen inherentemente un valor bololeano: *True o False*
* Cualquier numero distinto de 0 o cualquier objeto no vacio tiene valor *True*.
* El numero 0, objetos vacios y el objeto especial *None* tienen valor *False*

```python
a =-10 #True
b = None #False

if a:
    print("a es True")

if not b:
    print("b es False")
```

```python
c = [1,2,3]
# a c le reasigno None
c = None #0 es igual a vacio a nada
if c:
    print("Lista no vacía")
else:
    print("Lista vacía")
```

### 3 Iteracion (bucles)

* Repeticion de un bloque de codigo
* La terminacion del bucle depende del tipo de bucle
* Dos tipos: *While* y *For*

#### Bucle 'While'

* Reoeticion de un bloque de codigo hasta que se deje de cumplir una expresion (es decir, hasta que una condicion evalue a *False*)
* Si la condicion evalua a *False* desde el principio, el bloque de codigo nunca se ejecuta
* Cuidado con los bucles infinitos

Desarrollar un programa en python donde el usuario ingrese dos numeros por teclado el primer numero va a ser el inicio de un rango el segundo numero va a ser el final de dicho rango mediante el ciclo while verificar cuantos numeros son pares y mostrar los numeros y el total de numero pares 

```python
inicio = int(input('Ingresar primer numero(inicio): '))
fin = int(input('Ingresar segundo numero(fin): '))

contador_p = 0
indice = inicio

while indice <= fin:
    if indice % 2 == 0:
        print (indice)
        contador_p += 1
    indice += 1
    
print ('Total de numeros pares: ', contador_p)
```

Desarrolle un programa en python donde permita al usuario ingresar un numero entero positivo repetitivamente el programa debe continuar mientras se ingrese un numero mayor a cero. Al finalizar el proceso debe mostrarme la suma de los numero ingresados 

```python
N1 = int(input('Ingresar numero'))
st = 0
while N1 > 0:
    st += N1
    N1 = int(input('Ingresar numero:'))
print('La suma total es: ', st)
```

Crear un programa en python donde el usuario ingrese un numero por tecalo verificar si un numero es primo usando whyle

```python
n = int(input('Ingresar numero'))
cd = 0

while n < n:
    n % n == 0
    cd += cd
    if cd < 0:
        print ('Es numero primo')        
    else :
        print ('No es numero primo', n)
```
