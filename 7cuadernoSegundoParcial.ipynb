### 1 Estructuras de control
#### 1.1 Seleccion (sentencias condicionales)
* Seleccion de una de varias alternativas en base a alguna condicion
* Indentación para estructurar el código
* Importante el símbolo ‘:’

```python
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente

```python
x = 3
y = [1, 2, 3]

print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

### Valo Booleano

* Todos los objetos en python tienen inherentemente un valor bololeano: *True o False*
* Cualquier numero distinto de 0 o cualquier objeto no vacio tiene valor *True*.
* El numero 0, objetos vacios y el objeto especial *None* tienen valor *False*

```python
a =-10 #True
b = None #False

if a:
    print("a es True")

if not b:
    print("b es False")
```

### 3 Iteracion (bucles)

* Repeticion de un bloque de codigo
* La terminacion del bucle depende del tipo de bucle
* Dos tipos: *While* y *For*

#### Bucle 'While'

* Reoeticion de un bloque de codigo hasta que se deje de cumplir una expresion (es decir, hasta que una condicion evalue a *False*)
* Si la condicion evalua a *False* desde el principio, el bloque de codigo nunca se ejecuta
* Cuidado con los bucles infinitos

 Formato general<br>
  while test: #Minetras se cumpla la condicion<br>
  statements: #Instrucciones a ejecutar

```python
# # Ejemplo: Mostrar los primeros 3 objetos de una lista

indice = 0
numeros = [9, 4, 7, 1, 2]

while indice < 3:
    print(numeros[indice])
    indice =indice+1
```

```python
# Ejemplo: contar y mostrar los números inferiores a 10.

numeros = [33, 3, 9, 21, 1, 7, 12, 10, 8]
contador = 0
indice = 0
while indice < len(numeros):
    if numeros[indice] < 10:
        print(numeros[indice])
        contador += 1
    indice += 1
print('Contador:', contador)
print('Indice:', indice)
```

```python
nombre = 'Pablo'
while nombre: # Mientras 'nombre' no sea vacío
    print(nombre)
    nombre = nombre[1:]
```

```python
# #bucle infinito
i = 0
while i < 10:
    print(i)
    i += 1
```

Desarrollar un programa en python donde el usuario ingrese dos numeros por teclado el primer numero va a ser el inicio de un rango el segundo numero va a ser el final de dicho rango mediante el ciclo while verificar cuantos numeros son pares y mostrar los numeros y el total de numero pares 

```python
inicio = int(input('Ingresar primer numero(inicio): '))
fin = int(input('Ingresar segundo numero(fin): '))

contador_p = 0
indice = inicio

while indice <= fin:
    if indice % 2 == 0:
        print (indice)
        contador_p += 1
    indice += 1
    
print ('Total de numeros pares: ', contador_p)
```

Desarrolle un programa en python donde permita al usuario ingresar un numero entero positivo repetitivamente el programa debe continuar mientras se ingrese un numero mayor a cero. Al finalizar el proceso debe mostrarme la suma de los numero ingresados

Crear un programa en python donde el usuario ingrese un numero por tecalo verificar si un numero es primo usando whyle

```python
n = int(input('Ingresar numero'))
divisor = 1
cd = 0

while divisor <= n:
    if n % divisor == 0:
        cd += 1
    divisor += 1
    
if cd == 2:
    print (f'El numero {n} es primo')
else :
    print (f'El numero {n} no es primo. Tiene {cd} divisores')
```

### Bucle for 

* Permite recorrer los items de una sequencia o un objeto iterable.
* Funciona en strings, listas, tuplas, etc.

Formato general<br>
for item in objeto: #Asigna los items del objeto a la variable de cada item en cada iterracion<br>
statements #Instrucciones a ejecutar

```python
peliculas = ['Matrix', 'The purge', 'Avatar', 'Star Wars']
for pelicula in peliculas:
    print(pelicula)
```

* También se usa para iterar un número preestablecido de veces *(counted loops)*:

```python
for i in range(10):
    print(i)
```

* *range* es útil en combinación con len porque permite acceder a los elementos de una secuencia
por posición.

```python
nombre = 'Pablo'
for i in range(len(nombre)):
    print(nombre[i])
```

```python
for letra in nombre:
    print(letra)
```

* Iterracion de tuplas

```python
tuplas = [(1, 2,4), (3, 4), (5, 6)]
for a, b,c in tuplas:
    print(a, b,c)
```

* Iterracion de diccionarios. Se iteran las claves

```python
diccionario = {'a': 1, 'b': 2, 'c': 3}
for key in diccionario:
    print(f"{key} => {diccionario[key]}")
```

* Para iterar los pares (clave-valor) o únicamente los valores, se deben usar los métodos items
y values.

```python
for key, value in diccionario.items():
    print(f"{key} => {value}")
```

```python
for value in diccionario.values():
    print(value)
```

* La variable *item* en la cabecera del *for* puede ser cualquier exprecion que sea valida como parte izquierda de una asignacion convencional

```python
for a, b, c in [(1, 2, 3), (4, 5, 6)]:
    print(a, b, c)
```

Desarrollar un programa en python donde el usuario ingrese un numero y genere de acuerdo al numero ingresado una lista, ver que numero es par o impar 

```python
n = int(input('Ingresar numero: '))
for i in range(n+1):
    if i % 2 == 0:
        print (f'El numero {i} encontrado es par')
    else:
        print (f'El numero {i} encontrado es impar')
```
