### 1 Estructuras de control
#### 1.1 Seleccion (sentencias condicionales)
* Seleccion de una de varias alternativas en base a alguna condicion
* Indentación para estructurar el código
* Importante el símbolo ‘:’

```python
x = int(input("Ingrese el valor de x"))
#Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente

```python
x = 3
y = [1, 2, 3]

print(x > 0)
print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

* Se pueden introducir mas 'ramas' en sentencias condicionales a traves de la palabra clave *elif*

```python
#Desarrolar un programa en python donde se ingrese un numero verificar si el 
#numero ingresado es positivo, negativo o nulo
x =-3
if x > 0:
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

* Anidamiento

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
```

### 2 Switch


```python
a =-6
match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
```

```python
a = input() #week day

match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
```

#### Expresion ternaria

* Estructura if-else condensado en una linea
* Se recomienda usor solo en casos sensillos

```python
# # # Ejemplo: cálculo del valor absoluto de un número.

x =-3
resultado = x if x >= 0 else-x
print(resultado)
```

```python
val = -3
if val >=0:
    resultado = val
else:
    resultado =-val

print(resultado)
```

#### Comparacion de variables : ‘==’ vs ‘is’

*  ‘==’ compara si el valor de las variables es el mismo
*  ‘is’ compara si los objetos en las variables son iguales (referencia)

```python
a = 1000
b = a

print(a is b)
print(a == b)

c=type(a)
print(c)
d=type(b)
print(d)

print(c==d)

print(id(a))
print(id(b))
```

```python
a = [1, 2, 3]
b = [1, 2, 3]
c = a

print(a is b)
print(a is c)
print(a == b)

print(id(a))
print(id(b))
print(id(c))
```
