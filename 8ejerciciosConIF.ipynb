1. Ingresar dos valores enteros por teclado y muestre por pantalla su division si el divisor es cero el programa debe presentar un error

```python
N1 = int(input('Ingresar N1: '))
N2 = int(input('Ingresar N2: '))

if N2 == 0:
    print('ERROR no se puede dividir para cero', N2)
else:
    R = N1 / N2
    print('El divisor no es cero', N2)
    print ('El resultado de la division es: ', R)
```

3. Escribrir un programa para una empresa que tiene salas de juegos para todas las edades y quiere calcular de forma automatica el precio que debe cobrar a sus clientes por la entrada. El programa debe preguntar al usuario el nombre y la edad del cliente y mostrar el precio de entrada si el cliente es menor de 4 años puede entrar gratis, si tiene entre 4 y 18 años debe pagar 5 dolares y si es mayor a 18 años debe pagar 10 dolares 

```python
Cliente = input('Ingresa tu nombre: ')
Edad = int(input('Ingresa tu edad: '))
if Edad <= 4:
    print (f'Bienvenido {Cliente}, puedes entrar GRATIS por tener {Edad} años. ' )
elif 4 <= Edad <= 18:
    print (f'Bienvenido {Cliente}, tu edad es {Edad} años. Para continuar antes debes pagar 5 dolares' )
else:
    print (f'Bienvenido {Cliente}, tu edad es {Edad} años. Para continuar antes debes pagar 10 dolares' )
```

4. Para atributar un determinado impuesto se debe ser mayor a 16 años y tener un ingreso iguales o superiores a mil dolares mensuales escribir un programa que pregunte al usuario su edad y sus ingresos mensuales y muestre por pantalla si el usuario debe atributar o no

5. Escribir un programa en python donde el usuario ingrese por teclado el nombre de la operacion suma, resta, multiplicacion o division, tambien el usuario debera ingresar 2 numeros de cualquier tipo mediante un swicth comparar que operacion ingresa y mostrara en pantalla el resultado
